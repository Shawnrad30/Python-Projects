### Web Scraping and Regular Expression

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
url = 'http://www.analytictech.com/mb021/mlk.htm'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

mlkj_speech = soup.find_all('p')

speech_combined = [p.text for p in mlkj_speech]

string_speech = ' '.join(speech_combined)

string_speech_cleaned = string_speech.replace('\r\n',' ')

speech_no_punctuation = re.sub(r'[^\w\s]', '', string_speech_cleaned)

speech_lower = speech_no_punctuation.lower()

speech_broken_out = re.split(r'\s+', speech_lower)

df = pd.DataFrame(speech_broken_out).value_counts()

df.to_csv(r'C:\Users\shawn\OneDrive\Documents\Data Analyst Courses\Python Programming for Beginners\Python Projects\Project 5\MLKJ_Speech_Counts.csv', header = ['Counts'], index_label = 'Word')
```
