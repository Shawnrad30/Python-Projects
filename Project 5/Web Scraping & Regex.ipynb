#### **Web Scraping & Regular Expressions**

This project is a beginner-friendly Python program that focuses on extracting and analyzing text data from a webpage using web scraping and regular expressions.

The ensuing program allows a user to:
- Connect to a webpage containing a full speech by Martin Luther King Jr.
- Extract all paragraph text from the page
- Combine and clean the text data
- Remove punctuation and standardize the text format
- Break the text into individual words
- Count the frequency of each word
- Export the results to a CSV file for further analysis

This project demonstrates:
- Sending web requests with `requests`
- Parsing HTML content with `BeautifulSoup`
- Cleaning and transforming text data
- Using regular expressions (`re`) for pattern matching and text processing
- Working with lists and strings in Python
- Performing word frequency analysis with `pandas`
- Exporting structured data to a CSV file

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import os
```

```python
base_path = os.getcwd()
file_path = os.path.join(base_path, "MLKJ_Speech_Counts.csv")
```

```python
url = 'http://www.analytictech.com/mb021/mlk.htm'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

mlkj_speech = soup.find_all('p')

speech_combined = [p.text for p in mlkj_speech]

string_speech = ' '.join(speech_combined)

string_speech_cleaned = string_speech.replace('\r\n',' ')

speech_no_punctuation = re.sub(r'[^\w\s]', '', string_speech_cleaned)

speech_lower = speech_no_punctuation.lower()

speech_broken_out = re.split(r'\s+', speech_lower)

df = pd.DataFrame(speech_broken_out).value_counts()

df.to_csv(file_path, header=['Counts'], index_label='Word')
```
