#### **Automatic File Sorter**

This project is a beginner-friendly Python program that focuses on organizing files within a directory based on their file type.

The ensuing program allows a user to:
- Scan a specified folder for files
- Automatically create subfolders for different file types (CSV, TSV, text, images)
- Move files into their corresponding folders based on file extension
- Keep the directory organized without manual sorting

This project demonstrates:
- Working with the `os` module for file and directory operations
- Using the `shutil` module to move files
- Creating directories programmatically
- Looping through files in a directory
- Using conditional logic to categorize files by extension

```python
import os, shutil
```

```python
path = os.getcwd() + "\\"

folder_names = ['CSV Files', 'TSV Files', 'Text Files', 'Image Files']

for folder in folder_names:
    if not os.path.exists(path + folder):
        os.makedirs(path + folder)

file_names = os.listdir(path)

for file in file_names:
    if ".csv" in file and not os.path.exists(path + "CSV Files\\" + file):
        shutil.move(path + file, path + "CSV Files\\" + file)
    elif ".tsv" in file and not os.path.exists(path + "TSV Files\\" + file):
        shutil.move(path + file, path + "TSV Files\\" + file)
    elif ".png" in file and not os.path.exists(path + "Image Files\\" + file):
        shutil.move(path + file, path + "Image Files\\" + file)
    elif ".txt" in file and not os.path.exists(path + "Text Files\\" + file):
        shutil.move(path + file, path + "Text Files\\" + file)
```
