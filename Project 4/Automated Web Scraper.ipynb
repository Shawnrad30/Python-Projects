#### **Automated Web Scraper**

This project is a beginner-friendly Python program that focuses on collecting cryptocurrency data from a website and storing the results in a CSV file.

The ensuing program allows a user to:
- Connect to a cryptocurrency webpage
- Pull the current name and price of Bitcoin
- Store the extracted data in a pandas DataFrame
- Add a timestamp for when the data was collected
- Save the results to a CSV file
- Repeat the process automatically at scheduled intervals

This project demonstrates:
- Importing and using external Python libraries
- Sending web requests with `requests`
- Parsing HTML content with `BeautifulSoup`
- Creating and exporting DataFrames with `pandas`
- Working with timestamps using `datetime`
- Automating repeated tasks with a `while` loop and `time.sleep()`

```python
import os
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
import time
```

```python
base_path = os.getcwd()
folder_path = os.path.join(base_path, "Crypto Web Puller")
os.makedirs(folder_path, exist_ok=True)

file_path = os.path.join(folder_path, "Crypto_Automated_Pull.csv")
```

```python
def automated_crypto_pull():
    url = 'https://coinmarketcap.com/currencies/bitcoin/'

    page = requests.get(url)

    soup = BeautifulSoup(page.text, 'html')

    crypto_name = soup.find('span', class_ = 'sc-c1554bc0-0 hhUhBo').text

    crypto_price = soup.find('span', class_ = 'sc-c1554bc0-0 RbQXx base-text').text

    final_crypto_price = crypto_price.replace('$','')

    date_time = datetime.now()

    dict = {'Crypto Name': crypto_name,
            'Price': final_crypto_price,
            'Timestamp': date_time}

    df = pd.DataFrame([dict])

    if os.path.exists(file_path):
        df.to_csv(file_path, mode = 'a', header = False, index = False)
    else:
        df.to_csv(file_path, index = False)

    print(df)
```

```python
while True:
    automated_crypto_pull()
    time.sleep(3600)
```
