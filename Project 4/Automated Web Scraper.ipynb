### Automated Web Scraper

```python
import os
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
import time
```

```python
def automated_crypto_pull():
    url = 'https://coinmarketcap.com/currencies/bitcoin/'

    page = requests.get(url)

    soup = BeautifulSoup(page.text, 'html')

    crypto_name = soup.find('span', class_ = 'sc-c1554bc0-0 hhUhBo').text

    crypto_price = soup.find('span', class_ = 'sc-c1554bc0-0 RbQXx base-text').text

    final_crypto_price = crypto_price.replace('$','')

    date_time = datetime.now()

    dict = {'Crypto Name': crypto_name,
            'Price': final_crypto_price,
            'Timestamp': date_time}

    df = pd.DataFrame([dict])

    if os.path.exists(r'C:\Users\shawn\OneDrive\Documents\Data Analyst Courses\Python Programming for Beginners\Python Projects\Project 4\Crypto Web Puller\Crypto_Automated_Pull.csv'):
        df.to_csv(r'C:\Users\shawn\OneDrive\Documents\Data Analyst Courses\Python Programming for Beginners\Python Projects\Project 4\Crypto Web Puller\Crypto_Automated_Pull.csv', mode = 'a', header = False, index = False)
    else:
        df.to_csv(r'C:\Users\shawn\OneDrive\Documents\Data Analyst Courses\Python Programming for Beginners\Python Projects\Project 4\Crypto Web Puller\Crypto_Automated_Pull.csv', index = False)
    print(df)
```

```python
while True:
    automated_crypto_pull()
    time.sleep(3600)
```
